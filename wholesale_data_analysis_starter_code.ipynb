## Part 2: Transform the Data
Do we know that this client spent the more money than client 66037? If not, how would we find out? Transform the data using the steps below to prepare it for analysis.

```python
# Create a column that calculates the 
# subtotal for each line using the unit_price
# and the qty
subtotal = df['unit_price'] * df['qty']
print(subtotal)

```

```python
# Create a column for shipping price.
# Assume a shipping price of $7 per pound
# for orders over 50 pounds and $10 per
# pound for items 50 pounds or under.

def calculate_shipping_price(row):
    shipping_weight = row["unit_weight"] * row["qty"]
    if shipping_weight > 50:
        return shipping_weight * 7
    else:
        return shipping_weight * 10
df["shipping_price"] = df.apply(calculate_shipping_price, axis=1)
df.head()

```

```python
# Create a column for the total price
# using the subtotal and the shipping price
# along with a sales tax of 9.25%
df['sales_tax_rate'] = 0.0925
df['total_price'] = (df['unit_price'] * df['qty'] + df['shipping_price']) * (1 + df['sales_tax_rate'])
df.head()
```

```python
# Create a column for the cost
# of each line using unit cost, qty, and
# shipping price (assume the shipping cost
# is exactly what is charged to the client).
df['cost'] = df['unit_cost'] * df['qty'] + df['shipping_price']
df.head()

```

```python
# Create a column for the profit of each line using line cost and line price
df['unit_profit'] = df['unit_price'] - df['unit_cost']
df['total_profit'] = df['unit_profit'] * df['qty']
df.head()

```

## Part 4: Summarize and Analyze
Use the new columns with confirmed values to find the following information.

```python
# How much did each of the top 5 clients by quantity
# spend? Check your work from Part 1 for client ids.
# Store the client ids of those top 5 clients in a list.
top_five_client_ids = [33615, 66037, 46820, 38378, 24741]

# Create an empty dictionary to store total spend for each client
total_spend_top_five_clients = {}

# Calculate total spend for each client
for client_id in top_five_client_ids:
    # Calculate total_price as unit_price * qty for each order
    client_orders = df[df["client_id"] == client_id]
    total_invoiced = (client_orders["unit_price"] * client_orders["qty"]).sum().round(2)
    total_spend_top_five_clients[client_id] = total_invoiced

# Print total spend for each client
for client_id, total_invoiced in total_spend_top_five_clients.items():
    print(f"Client ID: {client_id}, Total invoiced: ${total_invoiced:,.2f}")
```

```python
# Create a summary DataFrame showing the totals for the
# for the top 5 clients with the following information:
# total units purchased, total shipping price,
# total revenue, and total profit. Sort by total profit.
top_five_client_ids = [33615, 66037, 46820, 38378, 24741]

# Filter the DataFrame to include only the top 5 clients
top_clients_df = df[df['client_id'].isin(top_five_client_ids)]

# Group by 'client_id' and aggregate the required columns, excluding 'total_price' for total revenue
summary_df = top_clients_df.groupby('client_id').agg({
    'qty': 'sum',
    'shipping_price': 'sum',
    'total_profit': 'sum'
}).reset_index()

# Rename columns for clarity
summary_df = summary_df.rename(columns={
    'client_id': 'Client ID',
    'qty': 'Total Units Purchased',
    'shipping_price': 'Total Shipping Price',
    'total_profit': 'Total Profit'
})

# Sort the DataFrame by 'Total Profit' in descending order
summary_df.sort_values(by='Total Profit', ascending=False, inplace=True)

# Display the summary DataFrame
print(summary_df)
```

```python
# Format the data and rename the columns
# to names suitable for presentation.
# Currency should be in millions of dollars.
top_five_client_ids = [33615, 66037, 46820, 38378, 24741]

# Filter the DataFrame to include only the top 5 clients
top_clients_df = df[df['client_id'].isin(top_five_client_ids)]

# Group by 'client_id' and aggregate the required columns
aggregated_df = top_clients_df.groupby('client_id').agg({
    'qty': 'sum',
    'shipping_price': 'sum',
    'total_price': 'sum',
    'unit_cost': 'sum',  # Assuming 'unit_cost' needs to be summed; adjust if needed
    'total_profit': 'sum'
}).reset_index()

# Convert currency values to millions and format as strings with $ sign and M for million
currency_columns = ['shipping_price', 'total_price', 'unit_cost', 'total_profit']
for column in currency_columns:
    aggregated_df[column] = aggregated_df[column] / 1000000  # Convert to millions
    aggregated_df[column] = aggregated_df[column].map('${:,.2f}M'.format)  # Format as string

# Since the DataFrame is already aggregated and sorted by 'client_id', you may want to sort by 'total_profit' or another column
aggregated_df.sort_values(by='total_profit', ascending=False, inplace=True)

# Display the DataFrame
print(aggregated_df)

```

```python
#Sort the updated data by "Total Profit" form highest to lowest

# Filter the DataFrame to include only the top 5 clients
top_five_client_ids = [33615, 66037, 46820, 38378, 24741]
df_filtered = df[df['client_id'].isin(top_five_client_ids)]

# Aggregate data by client_id
aggregated_df = df_filtered.groupby('client_id').agg({
    'qty': 'sum',
    'shipping_price': 'sum',
    'total_price': 'sum',
    'unit_cost': 'sum',  # Adjust this if a different operation is needed for unit_cost
    'total_profit': 'sum'
}).reset_index()

# Sort the aggregated data by 'total_profit' from highest to lowest
sorted_aggregated_df = aggregated_df.sort_values(by='total_profit', ascending=False)

# Display the sorted DataFrame
print(sorted_aggregated_df)
```

```python
# Round the numeric columns to two decimal places
aggregated_df['shipping_price'] = aggregated_df['shipping_price'].round(2)
aggregated_df['total_price'] = aggregated_df['total_price'].round(2)
aggregated_df['unit_cost'] = aggregated_df['unit_cost'].round(2)
aggregated_df['total_profit'] = aggregated_df['total_profit'].round(2)

# Format the columns as currency
currency_columns = ['shipping_price', 'total_price', 'unit_cost', 'total_profit']
for column in currency_columns:
    aggregated_df[column] = aggregated_df[column].apply(lambda x: f"${x:,.2f}")

# Sort the aggregated data by 'total_profit' from highest to lowest
sorted_aggregated_df = aggregated_df.sort_values(by='total_profit', ascending=False)

# Display the sorted and formatted DataFrame
print(sorted_aggregated_df)
```
